### Save host_id - count, earning
Farem un scatterplot count vs. earning

```python
# df.sort_values(by=['calculated_host_listings_count'], ascending=False)
df_count = df.groupby(by='host_id').count()['id']
df_count.sort_values(ascending=False)
df_count = pd.DataFrame(df_count)

df['earnings'] = df['number_of_reviews'] * df['price'] * df['minimum_nights']
df_earning = df.groupby(by='host_id')['earnings'].sum()

df_group = pd.concat([df_count, df_earning], axis=1, join='inner')

df_group.to_csv('../data/output/count-id.csv')
df_group.head()
```
